```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ap = 21897
bp = -6538

Rf = 10000
E=3.3

app = ap * Rf
b = bp + app/(E*Rf) 


def calc_m(u,du,dRf):
    return np.abs(app/(Rf+dRf) * 1/(u+du) + b - app/((Rf+dRf)*E) - (app/(Rf+dRf) * 1/u + b - app/((Rf+dRf)*E)))

u = np.arange(1,3,0.05)
drf_list = np.arange(-9900,-9000,10)
u_x,drf_y = np.meshgrid(u,drf_list)
m = calc_m(u_x,0.001,drf_y)
plt.pcolormesh(u_x,drf_y,m)
plt.show()
```
